```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Download the 'punkt_tab' resource

import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def preprocess_plot(plot):
    plot = str(plot)
    tokens = word_tokenize(plot.lower())

    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]

    stop_words = set(stopwords.words('english'))

    filtered_words = [word for word in stripped if word not in stop_words]

    filtered_plot = ' '.join(filtered_words)
    return filtered_plot

films1['Plot'] = films1['Plot'].apply(preprocess_plot)
```

**Bag of Words**

```python
for index, row in films1.iterrows():
    words = ''
    for col in columns:
        words += ' '.join(str(row[col]).split()) + ' '
    films1.at[index, 'Bag_of_words'] = words.strip()

films1 = films1[['Film Title', 'Bag_of_words']]
```

```python
bag_of_words = films1['Bag_of_words'].tolist()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
count = CountVectorizer()
count_matrix = count.fit_transform(films1['Bag_of_words'])
count_matrix
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
cosine_sim
```

```python
indices = pd.Series(films1['Film Title'])
```

**Function for Movie Recommendation**

```python
def recommend(title, cosine_sim = cosine_sim):
    recommended_movies = []
    idx = indices[indices == title].index[0]
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)
    top_5_indices = list(score_series.iloc[1:6].index)

    for i in top_5_indices:
      if films1['Film Title'][i] not in recommended_movies:
        recommended_movies.append(list(films1['Film Title'])[i])

    return recommended_movies

recommend('Gone Girl')
```
